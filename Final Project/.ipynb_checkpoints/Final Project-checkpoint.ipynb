# Predicting YouTube Video Views

## Introduction
#### About the data set:
This dataset contains a list of features of YouTube videos that have trended from 10 different countries. Some of these features include the trending date, title, channel name, the video category, likes, dislikes, comment_count, and many others. We also chose to focus on only the U.S. data.

#### Problem:
How can we predict the views of a Youtube video based on features like the number of likes/dislikes, title, category, publish time, etc...? 

#### Hypothesis:
Videos with higher likes/dislikes, % capital letters, published in the evening, will have higher views compared to other trending videos. 

#### Method:
This problem obviously called for regression methods. We chose to use a standard linear regression model and a k Nearest Neighbors regression model in order to predict views.

```python
import pandas as pd
import sklearn
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
import numpy as np
```

```python
us_stats = pd.read_csv("USvideos.csv")
```

```python
us_stats.head()
```

```python
def upperPercent(S):
    return sum(1 for c in S if c.isupper())/len(S)
```

```python

us_stats = us_stats.drop('thumbnail_link', axis=1)
us_stats = us_stats.drop('video_id', axis=1)
us_stats['title'] = [upperPercent(S) for S in us_stats['title']]
us_stats['month'] = [int(str(s)[-2:]) for s in us_stats['trending_date']]
us_stats['year'] = [int(str(s)[:2]) for s in us_stats['trending_date']]
us_stats['day'] = [int(str(s)[3:5]) for s in us_stats['trending_date']]
us_stats = us_stats.drop('trending_date', axis=1)
us_stats['publish_hour'] = [int(str(s)[11:13]) for s in us_stats['publish_time']]
us_stats = us_stats.drop('publish_time', axis=1)
us_stats = us_stats.drop('description', axis=1)
us_stats['tags'] = [sum(1 for c in S if c=='|')+1 for S in us_stats['tags']]
us_stats = us_stats.drop('channel_title', axis=1)

sum_by_id = us_stats.groupby(['category_id'])['views'].sum().reset_index()
sum_by_id.columns = ["id", "views"]
sum_by_id = sum_by_id.sort_values('views', ascending=False).reset_index()
sum_by_id = sum_by_id.drop('index',axis=1)
us_stats['id_rank'] = [sum_by_id[sum_by_id['id'] == i].index[0]+1 for i in us_stats['category_id']]
us_stats.head()
```

```python
plt.matshow(us_stats.corr())
cb = plt.colorbar()
plt.show()
#print(us_stats.corr())
```

## Cleaning Data and Correlation Plot
We cleaned the data set to get more useable features, such as the percentage of capital letters in the title, the number of tags, publish times, and the rank of categories with the most views, and dropped features that were irrelevant or difficult to parse. We plotted a correlation matrix to help us pick which features to use in our regression models.

```python
plt.scatter(us_stats['category_id'], us_stats['views'])
plt.show()
```

```python
us_stats.head()
```

```python
X = us_stats[['title','category_id', 'tags', 'likes','dislikes','comment_count','comments_disabled','ratings_disabled']]
Y = us_stats[['views']]
```

```python
from sklearn.metrics import mean_squared_error

LRmodel = LinearRegression()
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.25, random_state=42)
LRmodel.fit(x_train,y_train)
LRpredictions = pd.DataFrame(LRmodel.predict(x_test))

print(LRmodel.score(x_train,y_train))
print(LRmodel.score(x_test,y_test))
```

```python
from sklearn.neighbors import KNeighborsRegressor

kNNmodel = KNeighborsRegressor(weights='distance')
kNNmodel.fit(x_train, y_train)
kNNpredictions = pd.DataFrame(kNNmodel.predict(x_test))

train = kNNmodel.score(x_train,y_train)
test = kNNmodel.score(x_test,y_test)

print(train)
print(test)
```

```python
points = 75
X_plt = [i for i in range(points)]
Y_plt = y_test[:points]

fig = plt.figure(figsize=(10,5))
ax = fig.add_axes()
plt.scatter(X_plt,Y_plt,color='navy',label='data')
plt.plot(X_plt,LRpredictions[:points],color='orange',label='Linear Regression prediction')
plt.plot(X_plt,kNNpredictions[:points],color='green',label='kNN Regression prediction')
plt.legend(loc = 'upper right')
plt.xlabel('Index of X data')
plt.ylabel('Views (1e7)')
plt.show()
```

## Model Comparison Graph
We plotted some values from the test set along with the corresponding predicted values from the two regression models. The predictions using the KNN Regression model fit the best, which implies that videos with similar traits often had similar amounts of views.

## Conclusion
Of the two prediction methods, kNN regression was clearly superior. Both the training and testing scores were higher compared to the linear regression model. This is probably because videos that are most similar among all criteria will have the most similar performance, compared to trying to predict a linear relationship betwen each data point and views.
